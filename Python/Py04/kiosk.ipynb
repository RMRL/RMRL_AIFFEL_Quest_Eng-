```python
class Shape:
    def __init__(self, base, top, height):
        self.base = base        # 밑변
        self.top = top          # 윗변
        self.height = height    # 높이

    # 평행사변형 넓이
    def parallelogram(self):
        return self.base * self.height

    # 사다리꼴 넓이
    def trapezoid(self):
        return (self.top + self.base) * self.height / 2

s = Shape(5, 3, 4)
print("평행사변형 넓이:", s.parallelogram())
print("사다리꼴 넓이:", s.trapezoid())

```

```python
coffee = ['americano', 'latte', 'mocha']
c_price = [2000, 3000, 3000]
non_coffee = ['yuza_tea', 'green_tea', 'choco_latte']
n_price = [2500, 2500, 3000]

class Kiosk:
    def __init__(self):
        self.menu = coffee + non_coffee
        self.price = c_price + n_price
        self.order_menu = []
        self.order_price = []

    def menu_print(self):
        for i in range(len(self.menu)):       # ← Tab 2번 (8칸) ✅
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    def menu_select(self):
        n = 0                                  # ← Tab 2번 (8칸) ✅
        while n < 1 or len(self.menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))
            if 1 <= n and n <= len(self.menu):
                self.price_sum = self.price[n-1]
                self.order_menu.append(self.menu[n-1])
                self.order_price.append(self.price[n-1])
                print(self.menu[n-1], ' : ', self.price[n-1], '원')
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    def pay(self):
        p = input("결제 방법 (Cash/1 또는 card/2): ")
        if p == 'Cash' or p == '1':
            print('I will call an employee.')
        elif p == 'card' or p == '2':
            print('Please insert the card aligning the IC chip.')
        else:
            print("잘못된 입력입니다.")

    def table(self):
        print('⟝' + '-' * 30 + '⟞')
        print('|' + ' ' * 31 + '|')
        print('상품명 :', self.order_menu)
        print('총 주문금액 :', self.price_sum, '원')
        print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')

a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()
```
