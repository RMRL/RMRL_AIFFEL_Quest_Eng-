```python
import random  # 랜덤 숫자 만드는 도구 가져오기

class Account:
    """
    은행 계좌 클래스
    예금주와 초기 잔액을 입력받아 계좌를 생성합니다
    """
    
    count = 0  # Q2: 클래스 변수 = 계좌 개수 (모든 계좌가 공유!)
    
    def __init__(self, owner, balance):
        # Q1: 생성자 = 계좌 만들 때 자동 실행!
        self.bank    = "SC은행"          # 은행이름 고정
        self.owner   = owner             # 예금주 이름 저장
        self.balance = balance           # 초기 잔액 저장
        self.account_num = (             # 계좌번호 랜덤 생성
            str(random.randint(100, 999)) + "-" +
            str(random.randint(10, 99))  + "-" +
            str(random.randint(100000, 999999))
        )
        self.deposit_count  = 0          # Q7: 입금 횟수 카운터
        self.deposit_log    = []         # Q10: 입금 내역 저장 리스트
        self.withdraw_log   = []         # Q10: 출금 내역 저장 리스트
        
        Account.count += 1              # Q2: 계좌 생성될 때마다 +1
    
    
    def get_account_num(self):
        # Q3: 클래스 변수 출력 메서드
        # 지금까지 만들어진 계좌 개수 출력
        print(f"현재 계좌 개수: {Account.count}개")
    
    
    def deposit(self, amount):
        # Q4: 입금 메서드
        # 최소 1원 이상만 입금 가능!
        if amount >= 1:                          # 1원 이상이면
            self.balance += amount               # 잔액에 더하기
            self.deposit_count += 1              # Q7: 입금 횟수 +1
            self.deposit_log.append(amount)      # Q10: 입금 내역 기록
            print(f"{amount}원 입금 완료!")
            print(f"현재 잔액: {self.balance:,}원")
            
            # Q7: 입금 5회마다 이자 1% 지급!
            if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01)   # 잔액의 1%
                self.balance += interest               # 잔액에 이자 추가
                print(f"이자 {interest:,}원 지급! (잔액의 1%)")
        
        else:                                    # 1원 미만이면
            print("입금은 최소 1원 이상이어야 합니다!")
    
    
    def withdraw(self, amount):
        # Q5: 출금 메서드
        # 잔고 이상으로 출금 불가!
        if amount <= self.balance:               # 잔액 이하면
            self.balance -= amount               # 잔액에서 빼기
            self.withdraw_log.append(amount)     # Q10: 출금 내역 기록
            print(f"{amount}원 출금 완료!")
            print(f"현재 잔액: {self.balance:,}원")
        
        else:                                    # 잔액 초과면
            print("잔액이 부족합니다!")
            print(f"현재 잔액: {self.balance:,}원")
    
    
    def display_info(self):
        # Q6: 계좌 정보 출력 메서드
        # 잔액은 세자리마다 쉼표 표시!
        print("=" * 30)
        print(f"은행이름: {self.bank}")
        print(f"예금주:   {self.owner}")
        print(f"계좌번호: {self.account_num}")
        print(f"잔고:     {self.balance:,}원")  # :, = 세자리마다 쉼표!
        print("=" * 30)
    
    
    def deposit_history(self):
        # Q10: 입금 내역 출력 메서드
        print("=" * 30)
        print("입금 내역")
        print("=" * 30)
        if len(self.deposit_log) == 0:           # 입금 내역 없으면
            print("입금 내역이 없습니다.")
        else:
            for i, amount in enumerate(self.deposit_log):
                print(f"{i+1}회: {amount:,}원 입금")
    
    
    def withdraw_history(self):
        # Q10: 출금 내역 출력 메서드
        print("=" * 30)
        print("출금 내역")
        print("=" * 30)
        if len(self.withdraw_log) == 0:          # 출금 내역 없으면
            print("출금 내역이 없습니다.")
        else:
            for i, amount in enumerate(self.withdraw_log):
                print(f"{i+1}회: {amount:,}원 출금")


# =============================================
# Q8: 여러 객체 생성 — 3개 이상 인스턴스!
# =============================================
a1 = Account("파이썬", 10000)   # 계좌 1번 생성
a2 = Account("아이펠", 50000)   # 계좌 2번 생성
a3 = Account("클로드", 30000)   # 계좌 3번 생성

accounts = [a1, a2, a3]         # 리스트에 저장!


# =============================================
# 테스트 실행!
# =============================================

# Q6: 계좌 정보 출력
a1.display_info()

# Q3: 계좌 개수 출력
a1.get_account_num()

# Q4: 입금 테스트
a1.deposit(5000)
a1.deposit(3000)
a1.deposit(2000)
a1.deposit(1000)
a1.deposit(4000)   # 5번째 입금 → 이자 자동 지급!

# Q5: 출금 테스트
a1.withdraw(1000)
a1.withdraw(999999)  # 잔액 초과 → 오류 메시지!

# Q10: 입출금 내역 출력
a1.deposit_history()
a1.withdraw_history()




```
