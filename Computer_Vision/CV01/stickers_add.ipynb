```python
import os
import cv2
import dlib
import matplotlib.pyplot as plt
import numpy as np

# [AI 준비 단계: 데이터 경로 설정]
base_path = '/home/jovyan/work/camera_sticker'
img_path = base_path + '/images/image.png'
sticker_path = base_path + '/images/king.png'
model_path = base_path + '/models/shape_predictor_68_face_landmarks.dat'

# 1. 원본 데이터 로드: 인공지능은 매번 '오염되지 않은' 순수 데이터를 입력받아야 합니다.
img_original = cv2.imread(img_path) 
img_sticker = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED) # Alpha(투명도) 채널 포함 로드
img_result = img_original.copy() # 시각화를 위한 출력용 버퍼 생성

# 2. 얼굴 검출(Face Detection): HOG 알고리즘을 사용해 이미지 내 '얼굴 영역'의 사각형 좌표를 찾습니다.
detector_hog = dlib.get_frontal_face_detector()
dlib_rects = detector_hog(img_original, 1)

# 3. 랜드마크 추출(Landmark Prediction): 머신러닝 모델이 얼굴 내 68개의 핵심 포인트(눈, 코, 입, 눈썹)를 예측합니다.
predictor = dlib.shape_predictor(model_path)

for dlib_rect in dlib_rects:
    # 얼굴 안의 특징점들을 추출하여 리스트로 변환합니다.
    shape = predictor(img_original, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), shape.parts()))
    
    # 4. 좌표 계산(Geometry Calculation): AI가 인지한 눈썹 사이 중앙(27번 점)을 기준으로 왕관 위치를 계산합니다.
    # 인공지능 좌표계는 상단이 0이며, 아래로 갈수록 Y값이 커집니다. 따라서 위로 올리려면 빼야(-) 합니다.
    cx, cy = list_points[27] # 눈썹 사이 미간 지점
    
    # 왕관 크기 최적화: 검출된 얼굴 사각형 폭의 80%로 왕관 크기를 리사이징합니다.
    w = int(dlib_rect.width() * 0.8)
    h = int(img_sticker.shape[0] * (w / img_sticker.shape[1]))
    sticker_resized = cv2.resize(img_sticker, (w, h))

    # [핵심 수식] 
    # refined_y = 미간 높이(cy) - 왕관 높이(h) - 오프셋(70)
    # 왕관의 '바닥'이 눈썹보다 훨씬 위(이마 위)에 오도록 마이너스 연산을 강력하게 적용합니다.
    rx = cx - w // 2
    ry = cy - h - 70 

    # 5. 경계 처리(Exception Handling): 인공지능이 계산한 좌표가 이미지 영역(0 미만)을 벗어나지 않도록 보정합니다.
    if ry < 0: ry = 0
    if rx < 0: rx = 0

    # 6. 알파 블렌딩(Alpha Blending): 투명도(Alpha)를 활용해 원본과 스티커를 수학적으로 합성합니다.
    # 스티커의 투명도 정보를 0~1 사이의 가중치로 변환합니다.
    sticker_rgb = sticker_resized[:, :, :3] # 색상 채널
    alpha_mask = sticker_resized[:, :, 3] / 255.0 # 투명도 채널

    # 합성할 대상 영역(ROI)을 원본에서 잘라냅니다.
    target_area = img_result[ry:ry+h, rx:rx+w]

    # 가중 합산 수식: (원본 * 투명할 확률) + (스티커 * 색이 있을 확률)
    for c in range(3):
        target_area[:, :, c] = (1.0 - alpha_mask) * target_area[:, :, c] + alpha_mask * sticker_rgb[:, :, c]

# 7. 최종 출력: BGR(OpenCV 방식)을 RGB(표준 방식)로 변환하여 아티스트에게 결과를 보여줍니다.
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB))
plt.axis('off') # 불필요한 좌표축 제거
plt.show()
```

```python
# [AI 마스터 리셋 코드] 
import cv2, dlib
import matplotlib.pyplot as plt

# 1. 재료 새로 가져오기 (리셋의 핵심)
img_clean = cv2.imread(img_path) 
img_sticker = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED)
img_draw = img_clean.copy() # 매번 낙서 없는 새 도화지 준비!

# 2. 얼굴 및 미간(27번 점) 찾기
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(model_path)
rects = detector(img_clean, 1)

for rect in rects:
    shape = predictor(img_clean, rect)
    points = list(map(lambda p: (p.x, p.y), shape.parts()))
    
    # 미간(27번)을 기준으로 잡아야 이마로 가기 쉽다.
    cx, cy = points[27] 
    
    w = int(rect.width() * 0.8)
    h = int(img_sticker.shape[0] * (w / img_sticker.shape[1]))
    st_resized = cv2.resize(img_sticker, (w, h))

    # 3. [좌표 보정] 숫자가 작아질수록 하늘로 올라간다!
    rx = cx - w // 2
    ry = cy - h - 100 # -100 정도면 머리 위 옥상에 안착

    # [디버깅] 실제 숫자가 변하는지 터미널에서 확인
    print(f"현재 계산된 왕관 높이(Y): {ry}")

    # 4. 합성 작업 (Alpha Blending)
    st_rgb = st_resized[:, :, :3]
    alpha = st_resized[:, :, 3] / 255.0
    
    # 이미지를 벗어나면 에러가 나므로, 안전하게 영역을 확보
    if ry < 0: ry = 0 # 너무 올라가면 천장에 붙임

    target = img_draw[ry:ry+h, rx:rx+w]
    for c in range(3):
        target[:, :, c] = (1.0 - alpha) * target[:, :, c] + alpha * st_rgb[:, :, c]

# 5. 결과 확인
plt.imshow(cv2.cvtColor(img_draw, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import os
import cv2
import dlib
import matplotlib.pyplot as plt
import numpy as np

# [1단계] 시스템 환경 변수 및 자원(Assets) 경로 정밀 정의
# AI 모델과 원천 데이터의 참조 무결성을 위해 경로 변수를 상수로 관리합니다.
base_path = '/home/jovyan/work/camera_sticker'
img_path = base_path + '/images/image.png'
# 아티스트님께서 확인해주신 정확한 파일명 'cat_whisker.png'를 반영합니다.
sticker_path = base_path + '/images/cat_whisker.png' 
model_path = base_path + '/models/shape_predictor_68_face_landmarks.dat'

# [2단계] 데이터 인화 및 리셋 (Data Integrity)
# 멱등성(Idempotency) 확보를 위해 매 실행 시 원본 이미지를 새로 로딩하여 메모리 오염을 방지합니다.
img_original = cv2.imread(img_path)
img_sticker = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED) # Alpha(투명도) 채널 포함 필수

# 런타임 안정성을 위한 예외 처리 (NoneType 에러 방어 로직)
if img_sticker is None:
    print(f"❌ 데이터 로드 실패: {sticker_path} 경로를 다시 확인해주세요.")
else:
    img_result = img_original.copy() 

    # [3단계] Face Detection: HOG 알고리즘 기반 얼굴 영역 인식
    # Histogram of Oriented Gradients 기법을 사용하여 이미지 내 얼굴의 Bounding Box를 검출합니다.
    detector_hog = dlib.get_frontal_face_detector()
    dlib_rects = detector_hog(img_original, 1)

    # [4단계] Facial Landmark Prediction: 68-Point 회귀 추론
    # 사전 학습된 앙상블 회귀 모델을 통해 안면의 기하학적 구조 68개 좌표를 예측합니다.
    predictor = dlib.shape_predictor(model_path)

    for dlib_rect in dlib_rects:
        # 특징점 추출 및 벡터화(Vectorization)
        shape = predictor(img_original, dlib_rect)
        list_points = list(map(lambda p: (p.x, p.y), shape.parts()))
        
        # [5단계] 기하학적 정렬 (Geometric Alignment)
        # 고양이 수염의 안착점인 비첨(Nose Tip, 30번 랜드마크)을 중심 앵커 포인트로 설정합니다.
        cx, cy = list_points[30] 
        
        # 동적 스케일링(Dynamic Scaling): 검출된 얼굴 너비에 맞춰 스티커 해상도를 최적화합니다.
        # 고양이 수염의 폭을 얼굴 폭의 90%로 설정하여 심미적 완성도를 높였습니다.
        w = int(dlib_rect.width() * 0.9) 
        h = int(img_sticker.shape[0] * (w / img_sticker.shape[1]))
        sticker_resized = cv2.resize(img_sticker, (w, h))

        # [좌표 보정 공식]
        # 스티커의 기하학적 중심이 코 끝 좌표(30번)와 일치하도록 보정 계수를 적용합니다.
        # 수염은 코 중심에 위치해야 하므로 스티커 높이와 너비의 절반(1/2)을 감산합니다.
        rx = cx - w // 2
        ry = cy - h // 2 

        # [6단계] 경계값 예외 처리 (Boundary Clipping)
        # 합성 영역이 행렬 인덱스 범위를 초과하지 않도록 보정하여 시스템 안정성을 확보합니다.
        if rx < 0: rx = 0
        if ry < 0: ry = 0
        if ry + h > img_original.shape[0]: h = img_original.shape[0] - ry
        if rx + w > img_original.shape[1]: w = img_original.shape[1] - rx

        # [7단계] Alpha Blending: 투명도 기반 선형 보간 합성
        # 전경(Foreground)과 배경(Background)을 수학적 가중치에 따라 융합합니다.
        sticker_rgb = sticker_resized[:h, :w, :3]
        alpha_mask = sticker_resized[:h, :w, 3] / 255.0

        # 합성 대상 관심 영역(ROI: Region of Interest) 추출
        target_area = img_result[ry:ry+h, rx:rx+w]

        # 벡터 연산 기반의 픽셀 데이터 합성 (Alpha Blending Formula 적용)
        for c in range(3):
            target_area[:, :, c] = (1.0 - alpha_mask) * target_area[:, :, c] + alpha_mask * sticker_rgb[:, :, c]

    # [8단계] 최종 결과 렌더링 및 시각화 (Visualization)
    plt.figure(figsize=(12, 12))
    plt.imshow(cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Advanced AI Synthesis: Cat Whisker Results", fontsize=15)
    plt.show()
    
```
